```python
df1.drop(columns=["Unnamed: 0"],inplace=True,axis=1)
df2.drop(columns=["Unnamed: 0"],inplace=True,axis=1)
df3.drop(columns=["Unnamed: 0"],inplace=True,axis=1)
```

```python
def convert_to_datetime(data, column):

  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

```python
df1=convert_to_datetime(df1,"timestamp")
df1.info()
```

```python
df2=convert_to_datetime(df2,"timestamp")
df2.info()
```

```python
df3=convert_to_datetime(df3,"timestamp")
df3.info()
```

```python
from datetime import datetime
def convert_to_hourly(data,column):
    dummy=data.copy()
    l=dummy[column].tolist()
    l=[i.strftime("%Y-%m-%d %H:00:00") for i in l]
    l=[datetime.strptime(i,"%Y-%m-%d %H:00:00") for i in l]
    dummy[column]=l
    return dummy
```

```python
df1=convert_to_hourly(df1,"timestamp")
df1.head()
```

```python
df2=convert_to_hourly(df2,"timestamp")
df2.head()
```

```python
df3=convert_to_hourly(df3,"timestamp")
df3.head()
```

```python
df1_agg=df1.groupby(['timestamp','product_id']).agg({"quantity":"sum"}).reset_index()
df1_agg.head()
```

```python
df2_agg=df2.groupby(['timestamp','product_id']).agg({"estimated_stock_pct":"mean"}).reset_index()
df2_agg.head()
```

```python
df3_agg=df3.groupby(['timestamp']).agg({"temperature":"mean"}).reset_index()
df3_agg.head()
```

```python
merged_df=df2_agg.merge(df1_agg,on=["timestamp","product_id"],how="left")
merged_df.head()
```

```python
merged_df=merged_df.merge(df3_agg,on=["timestamp"],how="left")
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df["quantity"]=merged_df["quantity"].fillna(0)
merged_df.info()
```

```python
product_categories=df1[["product_id","category"]]
product_categories=product_categories.drop_duplicates()
product_price=df1[["product_id","unit_price"]]
product_price=product_price.drop_duplicates()
```

```python
merged_df=merged_df.merge(product_categories,on=["product_id"],how="left")
merged_df=merged_df.merge(product_price,on=["product_id"],how="left")
merged_df.head()
```

```python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df=pd.get_dummies(merged_df,columns=["category"])
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df.drop(columns=["product_id"],inplace=True)
merged_df.head()
```
